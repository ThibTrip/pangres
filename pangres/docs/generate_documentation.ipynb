# Notebook to generate the documentation of pangres

This notebook uses my library [npdoc_to_md](https://github.com/ThibTrip/npdoc_to_md) to render markdown files (located in the same folder as this notebook) with Jinja like placeholders (docstrings of functions, methods or classes are pulled and converted to pretty markdown).

```python
import os
# pip install npdoc_to_md
from npdoc_to_md import render_md_file, get_markdown_files_in_dir
```

# Configuration

Replace wiki_path by its corresponding absolute path

```python
wiki_path = 'WIKI_PATH'
```

# Render the files where you want

```python
files_and_names = get_markdown_files_in_dir('.')

for file, name in files_and_names.items():
    # render the file and put it a given destination
    destination = os.path.join(wiki_path, name)
    render_md_file(source=file, destination=destination)
```
